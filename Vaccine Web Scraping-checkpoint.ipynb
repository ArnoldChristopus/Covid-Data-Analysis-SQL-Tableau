```python
import requests
from bs4 import BeautifulSoup as bs
import selenium
from selenium.webdriver.common.by import By
from selenium import webdriver
import pandas as pd
```

```python
url = 'https://www.worldometers.info/coronavirus/'
html = requests.get(url).text
soup = bs(html, 'html.parser')

table = soup.find('table')
tr = table.find_all('tr')
print(tr[0])
print('\n')
print(tr[1])
```

```python
cov_columns = [i.text for i in tr[0].find_all('th')]
cov_columns
```

```python
cov_data = []
for rows in tr:
    values = [value.text for value in rows.find_all('td')]
    cov_data.append(values)
```

```python
covid_df = pd.DataFrame(data = cov_data, columns = cov_columns)
covid_df.drop(list(covid_df.filter(regex = 'pop|ppl')), axis = 1, inplace = True)
covid_df
```

```python
covid_df.to_csv('CovidCases.csv', encoding='utf-8', index=False)
```

```python
option = webdriver.ChromeOptions()
option.add_argument('headless')   # This option hides the browser

driver = webdriver.Chrome("C:\\Users\\asus1\\Downloads\\chromedriver.exe", options=option)
driver.get('https://ourworldindata.org/covid-vaccinations?country=OWID_WRL')
driver.implicitly_wait(10)

driver.implicitly_wait(10)
table = driver.find_element(By.XPATH, '/html/body/main/article/div[3]/div[2]/div/div/section[16]/div[3]/div/div/table')

vac_columns = []
for value in table.find_elements(By.XPATH, './/tr')[0].find_elements(By.XPATH, './/th'):
    vac_columns.append(value.text)

vac_data = []
for rows in table.find_elements(By.XPATH, './/tr'):
    values = [value.text for value in rows.find_elements(By.XPATH, './/td')]
    vac_data.append(values)
    
driver.close()

print(vac_columns)
print(vac_data[0:5])
```

```python
vaccine_df = df = pd.DataFrame(data = vac_data, columns = vac_columns)
vaccine_df
```

```python
vaccine_df.to_csv('Vaccines.csv', encoding='utf-8', index=False)
```
